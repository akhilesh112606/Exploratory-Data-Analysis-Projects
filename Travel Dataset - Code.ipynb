```python
# Univariate analysis for Cateogriate Columns
plt.figure(figsize = (8, 5))
sns.countplot(x = "ProdTaken", data = df, palette = 'viridis')
plt.title("Uni-Variate Analysis")
plt.show()
```

```python
cats = [  # Categorical (string/object type)
    'TypeofContact',
    'Occupation',
    'Gender',
    'ProductPitched',
    'MaritalStatus',
    'Designation'
]

nums = [  # Numerical (int or float type)
    'CustomerID',
    'ProdTaken',
    'Age',
    'CityTier',
    'DurationOfPitch',
    'NumberOfPersonVisiting',
    'NumberOfFollowups',
    'PreferredPropertyStar',
    'NumberOfTrips',
    'Passport',
    'PitchSatisfactionScore',
    'OwnCar',
    'NumberOfChildrenVisiting',
    'MonthlyIncome'
]
```

```python
for cat_col in cats:
    plt.figure(figsize = (8, 5))
    sns.countplot(x = cat_col, data = df, palette = 'viridis')
    plt.title("Uni-Variate Analysis")
    plt.show()
```

```python
# Univariate Analysis for Numerical Column
```

```python
for num_column in nums:
    plt.figure(figsize = (8, 5))
    sns.histplot(df[num_column], kde = True)
    plt.title("Univariate Analaysis for Numerical")
    plt.show()
```
